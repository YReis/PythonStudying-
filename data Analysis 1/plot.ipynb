```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train.head(50)
```

```python
df_train.date= pd.to_datetime(df_train.date)
```

```python
df_train_monthly=df_train.resample(rule='M', on='date')['sales'].sum()
```

```python
df_train_monthly
```

```python
df_train_monthly_filtered =df_train_monthly[(df_train_monthly.index<"2015") & (df_train_monthly.index>"2013")] 
```

```python
df_train_monthly_filtered
```

```python
sns.lineplot(df_train_monthly_filtered)
```

```python
df_train.date=df_train.date.dt.to_period('M')
```

```python
df_train_multiple=df_train.groupby(by=['date', 'store_nbr','family']).sum().reset_index()
```

```python
df_train_2013=df_train_multiple[(df_train_multiple.date>='2013') & (df_train_multiple.date<'2015')]
```

```python
df_train_2013
```

```python
df_train_beverages = df_train_2013[df_train_2013.store_nbr.isin([1,2])]
```

```python
df_train_beverages
```

```python
df_train_beverages.date=df_train_beverages.date.dt.to_timestamp()
```

```python
df_train_beverages.head(10)
```

```python
sns.lineplot(data=df_train_beverages, x="date" , y="sales", hue='store_nbr')
```
